```python
import os
from dotenv import load_dotenv
#加载 .env文件中的环境变量,保存在代码当前路径下
load_dotenv()
import openai
from langchain_openai import OpenAI
from langchain.schema import AIMessage,SystemMessage,HumanMessage
from langchain_openai import ChatOpenAI
from langchain_openai import OpenAIEmbeddings
import json

llm= ChatOpenAI(model_name="chatglm3-6b",temperature=0.1)

import chromadb
from chromadb.utils import embedding_functions
# 指定chromadb提供的embedding function使用openai接口能力
bge_embeddings = embedding_functions.OpenAIEmbeddingFunction(
    api_base=os.environ["OPENAI_API_BASE"],
    api_key=os.environ["OPENAI_API_KEY"],
    model_name="bge-large-zh-1.5"
)
# 百度embedding接口
from langchain_wenxin.embeddings import WenxinEmbeddings
wx_embeddings = WenxinEmbeddings(truncate="START")
```

```python
sentence1=result.content
#sentence1='embedding 技术的发展可以追溯到 20 世纪五六十年代的语言学研究，其中最著名的是 Harris 在 1954 年提出的分布式语义理论（distributional semantic theory）。这个理论认为，单词的语义可以通过它们在上下文中的分布来表示，也就是说，单词的含义可以从其周围的词语中推断出来。'
sentence2='从2010年以来，随着深度学习技术的发展，embedding 技术得到了广泛的应用和研究。在这个时期，出现了一些重要的嵌入算法，例如Word2Vec、GloVe和FastText等。这些算法可以通过训练神经网络或使用矩阵分解等技术来学习单词的嵌入表示，使得判断语义相似度成为可能，开启了自然语言预训练的序章。'
sentence3='基于内容的 embedding 方法：这部分主要涉及到 NLP 相关的文本 embedidng 方法，包括静态向量 embedding（如 word2vec、GloVe 和 FastText）和动态向量 embedding（如 ELMo、GPT 和 BERT）'
sentence4='基于 Graph 的 embedding 方法：这部分主要是介绍图数据的 embedding 方法，包括浅层图模型（如 DeepWalk、Node2vec 和 Metapath2vec）和深度图模型（如基于谱的 GCN 和基于空间的 GraphSAGE）'
embedding_1st = bge_embeddings(sentence1)
embedding_2nd = bge_embeddings(sentence2)
embedding_3rd = bge_embeddings(sentence3)
embedding_4th = bge_embeddings(sentence4)
print(embedding_1st[0])
```

```python
import numpy as np
from numpy.linalg import norm
from ltp import StnSplit

# Calculate cosine similarity
similarity = np.dot(embedding_1st[0],embedding_2nd[0])/(norm(embedding_1st[0])*norm(embedding_2nd[0]))
# Convert to cosine distance
distance = 1 - similarity
print("s1 vs s2:", similarity,distance)
#cosine_similarity
similarity = np.dot(embedding_1st[0],embedding_3rd[0])/(norm(embedding_1st[0])*norm(embedding_3rd[0]))
print("s1 vs s3:", similarity)
similarity = np.array(embedding_1st[0]) @ np.array(embedding_3rd[0]).T
print("s1 vs s3:", similarity)
similarity = np.array(embedding_3rd[0]) @ np.array(embedding_4th[0]).T
print("s3 vs s4:", similarity)
```
